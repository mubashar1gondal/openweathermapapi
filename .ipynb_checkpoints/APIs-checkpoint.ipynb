```python
def make_dict(pokemon_name):
    data = r.get(f'https://pokeapi.co/api/v2/pokemon/{pokemon_name.lower()}').json()
    poke_dict = {
        'name': data['name'],
        'weight': data['weight'],
        'abilities': [ability_dict['ability']['name'] for ability_dict in data['abilities']],
        'types': [x['type']['name'] for x in data['types']]
    }
    return poke_dict
```

```python
class Pokemon:
    def __init__(self, pokemon_name):
        d = r.get(f'https://pokeapi.co/api/v2/pokemon/{pokemon_name.lower()}').json()
        self.name = d['name']
        self.weight = d['weight']
        self.abilities = [ability_dict['ability']['name'] for ability_dict in d['abilities']]
        self.types = [x['type']['name'] for x in d['types']]
        
```

```python
rattata = Pokemon('rattata')
raichu = Pokemon('raichu')
raichu.types
```

```python
pokemon = ['raichu', 'rattata', 'squirtle', 'charmander', 'bulbasaur', 'pikachu']
pokemon = [make_dict(x) for x in pokemon]
pokemon
```
